## Data Preparation and Merging

### Subtask:
Select relevant economic indicators from both datasets that align with the hypothesis question. Clean the data by handling missing values (e.g., imputation, removal) and inconsistencies. Then, merge the GDP growth data with the selected economic indicators based on common identifiers such as country and year.


**Reasoning**:
I will start by cleaning and preparing the `df_gdp_growth` DataFrame as per the first instruction. This involves renaming a column, dropping rows with missing GDP growth rates, selecting specific columns, and renaming the 'Entity' column to 'Country' for standardization.



```python
import pandas as pd

# 1. Clean and Prepare df_gdp_growth
# Rename the GDP growth column
df_gdp_growth = df_gdp_growth.rename(columns={'Gross domestic product, constant prices - Percent change - Observations': 'GDP_Growth_Rate'})

# Drop rows where 'GDP_Growth_Rate' is missing
df_gdp_growth.dropna(subset=['GDP_Growth_Rate'], inplace=True)

# Select only relevant columns
df_gdp_growth = df_gdp_growth[['Entity', 'Year', 'GDP_Growth_Rate']]

# Rename 'Entity' to 'Country'
df_gdp_growth = df_gdp_growth.rename(columns={'Entity': 'Country'})

print("Cleaned df_gdp_growth head:")
print(df_gdp_growth.head())
print("\nCleaned df_gdp_growth info:")
df_gdp_growth.info()
```

**Reasoning**:
Now I will reshape and clean the `df_wdi` DataFrame as per the instructions. This involves filtering for annual data, dropping irrelevant metadata columns, renaming the country column, melting the year columns into 'Year' and 'Value', converting 'Year' to integer, and dropping rows with missing values in 'Value'.



```python
import pandas as pd

# 2. Reshape and Clean df_wdi
# Filter df_wdi to include only annual data where 'FREQ' is 'A'
df_wdi_filtered = df_wdi[df_wdi['FREQ'] == 'A'].copy()

# Drop metadata columns that are not useful for analysis or merging
drop_columns = [
    'FREQ', 'FREQ_LABEL', 'SEX', 'SEX_LABEL', 'AGE', 'AGE_LABEL',
    'UNIT_MEASURE', 'UNIT_MEASURE_LABEL', 'AGG_METHOD', 'AGG_METHOD_LABEL',
    'DECIMALS', 'DECIMALS_LABEL', 'DATABASE_ID', 'DATABASE_ID_LABEL',
    'UNIT_MULT', 'UNIT_MULT_LABEL', 'DATA_SOURCE', 'DATA_SOURCE_LABEL',
    'OBS_STATUS', 'OBS_STATUS_LABEL', 'OBS_CONF', 'OBS_CONF_LABEL'
]
df_wdi_filtered.drop(columns=drop_columns, inplace=True)

# Rename the 'REF_AREA_LABEL' column to 'Country'
df_wdi_filtered.rename(columns={'REF_AREA_LABEL': 'Country'}, inplace=True)

# Melt the DataFrame by transforming the year columns into 'Year' and 'Value'
year_columns = [col for col in df_wdi_filtered.columns if str(col).isdigit()]
id_vars = ['REF_AREA', 'Country', 'INDICATOR', 'INDICATOR_LABEL']
df_wdi_melted = df_wdi_filtered.melt(id_vars=id_vars, value_vars=year_columns, var_name='Year', value_name='Value')

# Convert the 'Year' column to an integer data type
df_wdi_melted['Year'] = pd.to_numeric(df_wdi_melted['Year'], errors='coerce').astype(int)

# Drop rows where the 'Value' column is missing (NaN)
df_wdi_melted.dropna(subset=['Value'], inplace=True)

print("Cleaned and Melted df_wdi head:")
print(df_wdi_melted.head())
print("\nCleaned and Melted df_wdi info:")
df_wdi_melted.info()
```

**Reasoning**:
I need to identify relevant economic indicators from the `df_wdi_melted` DataFrame. To do this, I will first print the unique values in the 'INDICATOR_LABEL' column to understand the available indicators and then define a list of key economic indicators based on these labels.



```python
import pandas as pd

# 3. Identify and Select Key Indicators
# Print unique values in 'INDICATOR_LABEL' to understand available indicators
print("Unique Indicator Labels from df_wdi_melted:")
print(df_wdi_melted['INDICATOR_LABEL'].unique())

# Based on the unique indicator labels, select a broad set of potentially influential indicators.
# This list can be expanded or refined as needed.
selected_indicators = [
    'GDP growth (annual %)',
    'Inflation, consumer prices (annual %)',
    'Population growth (annual %)',
    'Foreign direct investment, net inflows (% of GDP)',
    'Exports of goods and services (% of GDP)',
    'Imports of goods and services (% of GDP)',
    'Unemployment, total (% of total labor force) (modeled ILO estimate)',
    'Gross capital formation (% of GDP)',
    'Trade (% of GDP)',
    'Research and development expenditure (% of GDP)',
    'Fossil fuel energy consumption (% of total)',
    'Internet users (% of population)',
    'Electric power consumption (kWh per capita)',
    'Urban population (% of total population)'
]

# Filter df_wdi_melted to include only rows corresponding to these selected 'INDICATOR_LABEL' values
df_wdi_selected_indicators = df_wdi_melted[df_wdi_melted['INDICATOR_LABEL'].isin(selected_indicators)].copy()

print("\nFiltered df_wdi_selected_indicators head:")
print(df_wdi_selected_indicators.head())
print("\nFiltered df_wdi_selected_indicators info:")
df_wdi_selected_indicators.info()
print("\nUnique Indicator Labels in filtered DataFrame:")
print(df_wdi_selected_indicators['INDICATOR_LABEL'].unique())
```

**Reasoning**:
Now I will pivot the `df_wdi_selected_indicators` DataFrame. This will transform the unique 'INDICATOR_LABEL' values into separate columns, making the data suitable for merging with the GDP growth data. After pivoting, I will reset the index to convert 'Country' and 'Year' from index levels back into regular columns.



```python
import pandas as pd

# 4. Pivot Selected Indicators
# Pivot the filtered df_wdi_selected_indicators
# Each unique 'INDICATOR_LABEL' will become a new column
# 'Country' and 'Year' will form the index
df_wdi_pivoted = df_wdi_selected_indicators.pivot_table(
    index=['Country', 'Year'],
    columns='INDICATOR_LABEL',
    values='Value'
)

# Reset the index to make 'Country' and 'Year' regular columns
df_wdi_pivoted = df_wdi_pivoted.reset_index()

# Rename columns to remove special characters for easier access
df_wdi_pivoted.columns.name = None

# Display the first 5 rows of the pivoted DataFrame
print("\nPivoted df_wdi_pivoted head:")
print(df_wdi_pivoted.head())
print("\nPivoted df_wdi_pivoted info:")
df_wdi_pivoted.info()
```

**Reasoning**:
Now I will perform an inner merge of the prepared `df_gdp_growth` DataFrame with the `df_wdi_pivoted` DataFrame using 'Country' and 'Year' as common keys. This step combines the GDP growth data with the selected economic indicators into a single DataFrame for further analysis.



```python
import pandas as pd

# 5. Merge DataFrames
# Perform an inner merge of df_gdp_growth and df_wdi_pivoted on 'Country' and 'Year'
merged_df = pd.merge(df_gdp_growth, df_wdi_pivoted, on=['Country', 'Year'], how='inner')

# Display the first 5 rows of the merged DataFrame
print("\nMerged DataFrame head:")
print(merged_df.head())

# Display a summary of the merged DataFrame
print("\nMerged DataFrame info:")
merged_df.info()
```

## Correlation Analysis

### Subtask:
Calculate the correlation coefficients between GDP growth and the identified economic indicators to quantify the strength and direction of their relationships. This will help pinpoint factors that heavily influence the economy.


**Reasoning**:
I will start by dropping the redundant 'GDP growth (annual %)' column from `merged_df` to ensure there are no duplicate GDP growth indicators before performing correlation analysis.



```python
import pandas as pd

# 1. Drop the 'GDP growth (annual %)' column if it exists, as it is redundant with 'GDP_Growth_Rate'.
# This column was also present in the original WDI dataset and might have been merged.
if 'GDP growth (annual %)' in merged_df.columns:
    merged_df.drop(columns=['GDP growth (annual %)'], inplace=True)
    print("Dropped 'GDP growth (annual %)' column from merged_df.")
else:
    print("'GDP growth (annual %)' column not found in merged_df, skipping drop.")

print("Merged DataFrame after dropping redundant GDP column head:")
print(merged_df.head())
print("Merged DataFrame info after dropping redundant GDP column:")
merged_df.info()
```

**Reasoning**:
Now I will create a DataFrame containing only the numerical columns relevant for correlation analysis, handle missing values by dropping rows, calculate the Pearson correlation matrix, and then extract and display the correlation coefficients of 'GDP_Growth_Rate' with all other economic indicators, sorted by their absolute values.



```python
import pandas as pd

# 2. Create a new DataFrame containing only the numerical columns for correlation analysis
# Exclude 'Country' and 'Year' as they are identifiers, not numerical indicators for correlation
numerical_cols = merged_df.select_dtypes(include=['float64', 'int64']).columns.tolist()
df_correlation = merged_df[numerical_cols].copy()

# 3. Handle any remaining missing values in this new DataFrame by dropping rows
# Correlation calculations require complete data, so dropping NaNs is appropriate here.
df_correlation.dropna(inplace=True)

print("\nDataFrame for correlation analysis head (after dropping NaNs):")
print(df_correlation.head())
print("\nDataFrame for correlation analysis info (after dropping NaNs):")
df_correlation.info()

# 4. Calculate the Pearson correlation matrix for the cleaned numerical DataFrame
correlation_matrix = df_correlation.corr()

# 5. Extract and display the correlation coefficients of 'GDP_Growth_Rate' with all other economic indicators
gdp_growth_correlations = correlation_matrix['GDP_Growth_Rate'].drop('GDP_Growth_Rate')

# Sort in descending order of their absolute values
gdp_growth_correlations_sorted = gdp_growth_correlations.reindex(gdp_growth_correlations.abs().sort_values(ascending=False).index)

print("\nCorrelation of GDP_Growth_Rate with other economic indicators (sorted by absolute value):\n")
print(gdp_growth_correlations_sorted)
```

## Prepare Data for Basic Needs Analysis

### Subtask:
Filter `df_wdi_melted` for the selected basic needs indicators, pivot them to have indicators as columns, and merge with `df_imputed` using 'Country' and 'Year'. Ensure country codes (`REF_AREA`) are preserved for mapping.


**Reasoning**:
I will filter the `df_wdi_melted` DataFrame to include only rows corresponding to the `selected_basic_needs_indicators` and then pivot it to transform indicator labels into columns, which is a necessary step before merging.



```python
import pandas as pd

# 1. Filter the df_wdi_melted DataFrame for selected basic needs indicators
df_basic_needs_filtered = df_wdi_melted[df_wdi_melted['INDICATOR_LABEL'].isin(selected_basic_needs_indicators)].copy()

# 2. Pivot the filtered DataFrame
# 'REF_AREA', 'Country', 'Year' as index, 'INDICATOR_LABEL' as columns, 'Value' as values
df_basic_needs_pivoted = df_basic_needs_filtered.pivot_table(
    index=['REF_AREA', 'Country', 'Year'],
    columns='INDICATOR_LABEL',
    values='Value'
)

# 3. Reset the index to convert 'REF_AREA', 'Country', and 'Year' into regular columns
df_basic_needs_pivoted = df_basic_needs_pivoted.reset_index()

# Remove columns.name attribute for cleaner DataFrame operations
df_basic_needs_pivoted.columns.name = None

print("Pivoted Basic Needs DataFrame head:")
print(df_basic_needs_pivoted.head())
print("\nPivoted Basic Needs DataFrame info:")
df_basic_needs_pivoted.info()
```

**Reasoning**:
Now that the basic needs indicators have been pivoted, I will merge this DataFrame (`df_basic_needs_pivoted`) with the main imputed DataFrame (`df_imputed`) using 'Country' and 'Year' as common keys. After merging, I will impute any remaining missing numerical values in the newly merged DataFrame using the mean of each column to ensure completeness for further analysis. This step aligns with instructions 4 and 5 of the subtask.



```python
import pandas as pd

# 4. Merge the pivoted basic needs DataFrame with the df_imputed DataFrame
# We'll use a left merge to retain all rows from df_imputed and add basic needs data.
# 'REF_AREA' from df_basic_needs_pivoted might be useful later, so we keep it.
merged_df_basic_needs = pd.merge(
    df_imputed,
    df_basic_needs_pivoted,
    on=['Country', 'Year'],
    how='left' # Use left merge to keep all original GDP rows
)

# 5. Impute any remaining missing numerical values in the newly merged DataFrame by filling them with the mean of their respective columns
numerical_cols_final = merged_df_basic_needs.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Exclude 'Year' from mean imputation, as it's an identifier
if 'Year' in numerical_cols_final:
    numerical_cols_final.remove('Year')

# Perform mean imputation
for col in numerical_cols_final:
    if merged_df_basic_needs[col].isnull().any():
        merged_df_basic_needs[col] = merged_df_basic_needs[col].fillna(merged_df_basic_needs[col].mean())

# 6. Display the first few rows and a summary of the resulting DataFrame
print("\nMerged DataFrame with Basic Needs Indicators (head after imputation):")
print(merged_df_basic_needs.head())
print("\nMerged DataFrame with Basic Needs Indicators (info after imputation):")
merged_df_basic_needs.info()
```

## Identify Countries with Highest Need (Recent Year)

### Subtask:
For the most recent year with sufficient data, analyze the distribution of these indicators. Identify countries in the lowest percentiles for positive indicators (e.g., life expectancy) or highest percentiles for negative indicators (e.g., poverty, mortality rates).


**Reasoning**:
I need to identify the most recent year with sufficient data for the selected basic needs indicators. I will first determine the latest year present in the `merged_df_basic_needs` DataFrame, then check the non-null count for each `selected_basic_needs_indicators` column for that year to ensure data sufficiency before selecting the `most_recent_year`.



```python
import pandas as pd

# 1. Identify the most recent year with sufficient data

# Get the maximum year present in the merged_df_basic_needs
max_year = merged_df_basic_needs['Year'].max()

# Define a threshold for "sufficient data" (e.g., at least 25% non-null values for the selected indicators)
data_sufficiency_threshold_ratio = 0.25 # adjust as needed

most_recent_year = None
for year in range(max_year, merged_df_basic_needs['Year'].min() -1, -1):
    df_year = merged_df_basic_needs[merged_df_basic_needs['Year'] == year].copy()

    # Check if any data exists for the year
    if df_year.empty:
        continue

    # Calculate the number of non-null entries for the selected indicators in this year
    non_null_counts = df_year[selected_basic_needs_indicators].count().sum()

    # Calculate the total possible entries for the selected indicators in this year
    total_possible_entries = len(df_year) * len(selected_basic_needs_indicators)

    # If total_possible_entries is 0, skip to avoid division by zero
    if total_possible_entries == 0:
        continue

    # Check if the year has sufficient data based on the threshold
    if non_null_counts / total_possible_entries >= data_sufficiency_threshold_ratio:
        most_recent_year = year
        break

if most_recent_year:
    print(f"Most recent year with sufficient data for basic needs indicators: {most_recent_year}")
else:
    print("Could not find a recent year with sufficient data. Please adjust the sufficiency threshold or data range.")
    # Fallback to the latest year available if no year meets the threshold for all indicators,
    # but we will proceed with the available data for that year anyway.
    most_recent_year = merged_df_basic_needs['Year'].max()
    print(f"Defaulting to latest year: {most_recent_year}")

# 2. Filter merged_df_basic_needs to include only data for the most_recent_year
df_recent_year_needs = merged_df_basic_needs[merged_df_basic_needs['Year'] == most_recent_year].copy()

print(f"\nDataFrame for most recent year ({most_recent_year}) head:")
print(df_recent_year_needs.head())
print(f"\nDataFrame for most recent year ({most_recent_year}) info:")
df_recent_year_needs.info()
```

**Reasoning**:
Now that the `df_recent_year_needs` DataFrame has been prepared, I will define the `indicator_type` dictionary and then iterate through each selected basic needs indicator. For each indicator, I will filter out NaN values, sort the data appropriately based on whether 'high_is_bad' or 'low_is_bad' applies, and print the top 5 countries to identify those with the highest needs.



```python
import pandas as pd

# 3. Define the indicator_type dictionary
indicator_type = {
    'Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population)': 'high_is_bad',
    'Mortality rate, under-5 (per 1,000 live births)': 'high_is_bad',
    'Literacy rate, adult total (% of people ages 15 and above)': 'low_is_bad',
    'People using at least basic drinking water services (% of population)': 'low_is_bad',
    'People using at least basic sanitation services (% of population)': 'low_is_bad',
    'Government expenditure on education, total (% of GDP)': 'low_is_bad', # Lower expenditure can indicate higher need for investment
    'Prevalence of moderate or severe food insecurity in the population (%)': 'high_is_bad'
}

print(f"\nAnalysis for the most recent year: {most_recent_year}\n")

# 4. For each indicator, identify countries in the lowest/highest percentiles
for indicator, type_of_indicator in indicator_type.items():
    print(f"--- Indicator: {indicator} ---")

    # Filter out rows with NaN values for the current indicator
    df_filtered_indicator = df_recent_year_needs.dropna(subset=[indicator]).copy()

    if not df_filtered_indicator.empty:
        if type_of_indicator == 'high_is_bad':
            # Sort in descending order to find countries with highest need (highest value)
            top_countries = df_filtered_indicator.sort_values(by=indicator, ascending=False)
            print("Top 5 Countries with HIGHEST need (highest value):")
        elif type_of_indicator == 'low_is_bad':
            # Sort in ascending order to find countries with highest need (lowest value)
            top_countries = df_filtered_indicator.sort_values(by=indicator, ascending=True)
            print("Top 5 Countries with HIGHEST need (lowest value):")

        # Display the top 5 countries for the current indicator
        print(top_countries[['Country', indicator]].head(5).to_string(index=False))
    else:
        print("No data available for this indicator in the selected year after dropping NaNs.")
    print("\n")
```
